```python
# Import all necessary libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots/visualizations are displayed within output cells
%matplotlib inline
```

```python
# Load excel file
df = pd.read_csv("churn-bigml-80.csv")

# Output first 5 lines of file
df.head()
```

```python
# Explore the shape of the dataset (how many customers and features)
df.shape
```

```python
# Explore column names
df.columns
```

```python
# Analyze data types
df.info()
```

```python
# Show summary statistics
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Check distribution of churn
df['Churn'].value_counts()
```

```python
# Check percentage of churn
df['Churn'].value_counts(normalize=True)
```

Approximately 14.56% of customers churned, which indicates a moderately imbalanced dataset

```python
# Convert churn to numeric value
df['Churn_numeric'] = df['Churn'].astype(int)
```

```python
# Plot churn vs international plan
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='International plan', y='Churn_numeric')
plt.title("Churn Rate by International Plan")
plt.show()
```

```python
# Plot churn vs voice mail plan
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='Voice mail plan', y='Churn_numeric')
plt.title("Churn Rate by Voice Mail Plan")
plt.show()
```

```python
# Plot churn vs account length
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Churn', y='Account length')
plt.title("Account Length vs Churn")
plt.show()
```

```python
# Plot churn vs total day minutes
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Churn', y='Total day minutes')
plt.title("Total Day Minutes vs Churn")
plt.show()
```

```python
# Plot correlation heatmap for numeric values
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

- Customers with international plans show much higher churn rates
- Higher day minutes (and consequently charges) are associated with churn rates
- Account length is not a strong churn predictor
- Features based on usage (minutes/charges) and customer service calls are slightly more predictive than international call counts and voicemail messages
- Churn class is imbalanced
